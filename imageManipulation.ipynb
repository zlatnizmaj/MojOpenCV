```python
# use display() of IPython.display module and Image.fromarray() of PIL module
from PIL import Image
import cv2 
from IPython.display import display

img = cv2.imread('images/original3.jpg') # with the OpenCV function imread(), the order of colors is BGR (blue, green, red).

# In Pillow, the order of colors is assumed to be RGB (red, green, blue).
# As we are using Image.fromarray() of PIL module, we need to convert BGR to RGB.

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Converting BGR to RGB

display(Image.fromarray(img))
```

# 1. Resize Images


```python
# function resize()
for width in (100, 200, 400):
    resized = cv2.resize(original, (width, width), interpolation=cv2.INTER_LINEAR)
    print("Resized Image Width: ", resized.shape[0])
    cv2.imwrite(f"images/reszied_{width}.jpg", resized)
    
#  interpolation method for a smaller image is the linear interpolation 
# (i.e., INTER_LINEAR)
# desired method for a larger image is 
# an interpolation algorithm based on local areas (i.e., INTER_AREA)
```

# 2. Color to Gray

```python
# cvtColor
# Color to Grayscale
gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)

print(f"Gray Scale Image Shape: {gray.shape}")
print(f"Gray Scale Image Size: {gray.size}")
cv2.imwrite("original_gray.jpg", gray)
```
